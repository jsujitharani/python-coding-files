```python
import random
```

```python
#p1=random.randint(1,6)
#print(p1)
#p2=random.randint(1,6)
#print(p2)
snakes={40:20,50:16,81:78,95:36,92:52}
ladders={4:20,10:29,14:77,33:51,64:82,74:90}
player1_at=1
player2_at=1

while player1_at<100 and player2_at<100:
    
    input("Player 1, press Enter to roll the dice...")
    p1=random.randint(1,6)
    print("player1 rolled dice ",p1)
    player1_at+=p1
    if player1_at in snakes:
        print("this player 1 is at step",snakes[player1_at],"because of snake bite at ",player1_at)
        print("player1 is at step",snakes[player1_at])
        player1_at=snakes[player1_at]
    elif player1_at in ladders:
        print("this player 1 is at step",ladders[player1_at],"because of ladder",player1_at)
        print("player1 is at step",ladders[player1_at])
        player1_at=ladders[player1_at]
    elif player1_at>=100:
        player1_at=100
        print("player 1 winner")
        break
    else:
        print("player1 is at step",player1_at)
        
    
    input("Player 2, press Enter to roll the dice...")
    p2=random.randint(1,6)
    print("player2 rolled dice ",p2)
    player2_at+=p2
    if player2_at in snakes:
        print("this player 2 is at step",snakes[player2_at],"because of snake bite",player2_at)
        print("player2 is at step",snakes[player2_at])
        player2_at=snakes[player2_at]
    elif player2_at in ladders:
        print("this player 2 is at step",ladders[player2_at],"because f the ladder",player2_at)
        print("player2 is at step",ladders[player2_at])
        player2_at=ladders[player2_at]
    elif player2_at>=100:
        print("player 2 winner")
        break
    else:
        print("player2 is at step",player2_at)

#if player1_at>=100:
 #   print("player1 is the winner")
 #   break
#else:
 #   print("player2 is the winner")
  #  break
    
```
